```python
import pandas as pd
import numpy as np
import kaleido

import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
%matplotlib inline

pd.options.plotting.backend = "plotly"

import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

```python
df = pd.read_csv('tweets_neg_sample_perspective.csv', header=0)
df.columns = df.columns.map(lambda x: x.strip()[1:-1])
df.head()
```

```python
g_data = df.groupby('month').agg({
    'TOXICITY': np.median,
    'SEVERE_TOXICITY': np.median,
    'IDENTITY_ATTACK': np.median,
    'INSULT': np.median,
    'PROFANITY': np.median,
    'THREAT': np.median
})
g_data.head()
```

```python
figure = px.line(g_data, x=g_data.index, y=['TOXICITY', 'SEVERE_TOXICITY', 'IDENTITY_ATTACK', 'INSULT', 'PROFANITY', 'THREAT'])
figure.update_layout(
            title='Monthly Median Perspective Scores for Sample of Negative Tweets',
            xaxis_title = 'Month',
            yaxis_title = 'Monthly Median Score',
            legend_title = 'Legends')
figure.show()
```
