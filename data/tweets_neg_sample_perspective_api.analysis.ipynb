```python
import pandas as pd
import re
```

```python
df = pd.read_csv('all_tweets_neg_emo_outliers.tsv', sep='\t', header=None, names=['id', 'month', 'tweet'])
print(df.info())
print(df.head())
```

```python
print(df[df.tweet.isna()])
df.tweet = df.tweet.map(lambda x: re.sub(r'https?:\S+', ' ', x.strip()))
df.head()
```

```python
months = set(df.month.to_list())
print(f'Total months: {len(months)}')
```

```python
df.groupby("month").count()
```

```python
sample_df = pd.DataFrame()

for month in months:
    temp_df = df[df.month == month]
    total_tweets = temp_df.tweet.count()

    if total_tweets <= 1000:
        sample_df = sample_df.append(temp_df)
    else:
        sample_df = sample_df.append(temp_df.sample(n=1000, random_state=1))

print(f'Total tweets selected: {sample_df.groupby("month").count()}')
```

```python
print(sample_df.count())
sample_df.to_csv('tweets_neg_sample_perspective_api.tsv', sep='\t', index=False)
```
